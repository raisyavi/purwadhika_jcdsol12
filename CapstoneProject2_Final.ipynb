# Identification and Handling of Outliers:

Explanation:
- Utilizing the IQR (Interquartile Range) method to identify and address outliers in the 'Income' column.
- The steps involve calculating Q1, Q3, IQR, and determining upper and lower bounds to identify outliers.
- Outliers are addressed by replacing them with the upper or lower bound depending on specific conditions.
- Handling outliers is crucial to ensure more consistent and accurate results in analysis and model building.

```python
# Identifikasi dan penanganan outlier
def handle_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return data

# Handle outlier pada kolom 'Income'
df = handle_outliers(df, 'Income')
df['Income']
```

```python
from scipy.stats import ttest_ind
import numpy as np

# Mengubah kolom 'Education' menjadi dua kelompok: '<40' dan '>40'
df['Education_Group'] = np.where(df['Age'] < 40, '<40', '>40')

# Uji t independent untuk 'Income' antara dua kelompok
group1 = df[df['Education_Group'] == '<40']['Income']
group2 = df[df['Education_Group'] == '>40']['Income']

t_stat, p_value = ttest_ind(group1, group2)
print('Nilai ini adalah statistik uji t:')
print(t_stat)
print('Nilai ini adalah nilai p (probability):')
print(p_value)
```

Interpretasi Hasil:
Nilai Statistik Uji t:


1. Nilai statistik uji t sebesar -4.0363 menunjukkan seberapa jauh perbedaan rata-rata antara dua kelompok ('<40' dan '>40') dalam variabel 'Income'.

Nilai p (Probability):

2. Nilai p sangat kecil (0.0000561), jauh di bawah tingkat signifikansi umum (0.05). Hal ini menunjukkan bahwa terdapat perbedaan signifikan dalam rata-rata 'Income' antara dua kelompok tersebut.

```python
from scipy.stats import pearsonr

# Statistika Deskriptif
descriptive_stats = df[['Age', 'Income']].describe()
print("Statistika Deskriptif:")
print(descriptive_stats)

# Korelasi antara Age dan Income
correlation, p_value = pearsonr(df['Age'], df['Income'])
print("\nKorelasi antara Age dan Income:")
print(f"Korelasi Pearson: {correlation:.2f}")
print(f"P-Value: {p_value:.4f}")

# Jika ingin melihat scatter plot dengan garis regresi
plt.figure(figsize=(10, 6))
sns.regplot(x='Age', y='Income', data=df)
plt.title('Scatter Plot Age vs Income with Regression Line')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

```python
from scipy.stats import f_oneway

# Uji Anova untuk perbedaan signifikan antara kelompok pendidikan
anova_result = f_oneway(df[df['Education'] == 'PhD']['Income'],
                        df[df['Education'] == 'Master']['Income'],
                        df[df['Education'] == 'Graduation']['Income'],
                        df[df['Education'] == '2n Cycle']['Income'],
                        df[df['Education'] == 'Basic']['Income'])

print("ANOVA p-value:", anova_result.pvalue)
```

Hasil uji Anova menunjukkan bahwa ada perbedaan signifikan dalam distribusi 'Income' antara kelompok pendidikan (p-value < 0.05).
